```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['species']=iris.target
df['species'] = df['species'].map({0: 'setosa', 1:'versicolor', 2:'virginica'})
df.head()

df.info()

df.isnull().sum()

#Resumen estadistico
df.describe()

#dibujar histograma
df.drop('species', axis=1).hist(bins=20, figsize=(8, 6))
plt.suptitle('Distribucin de las Caractersticas del Conjunto de Datos Iris')
plt.show()

#graficar la distribucion de las especies
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df, palette='Set2')
plt.title('Distribucin de Especies')
plt.xlabel('Especie')
plt.ylabel('Frecuencia')
plt.show()

```
